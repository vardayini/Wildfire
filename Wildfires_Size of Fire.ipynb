```python
import sqlite3
import pandas as pd
```

```python
cnx = sqlite3.connect('FPA_FOD_20170508.sqlite')
```

```python
df = pd.read_sql_query("SELECT FIRE_YEAR,STAT_CAUSE_DESCR,LATITUDE,LONGITUDE,FIRE_SIZE FROM 'Fires'", cnx)
```

```python
df['STAT_CAUSE_DESCR'].value_counts()
```

```python
%matplotlib inline
```

```python
df_p = df[df['STAT_CAUSE_DESCR']=='Powerline']
```

```python
import plotly
plotly.offline.init_notebook_mode(connected=True)
```

```python
import plotly.express as px

fig = px.scatter_geo(df_p,  lon = 'LONGITUDE',
        lat = 'LATITUDE', color="STAT_CAUSE_DESCR",
                     hover_name="FIRE_SIZE", size="FIRE_SIZE",
                     projection="albers usa")
fig.show()
```
